## animate

```python

'''
Author: fengsc
Date: 2022-03-16 12:57:14
LastEditTime: 2022-03-16 14:47:19
'''
%reset -f
# !python版素质三连

%matplotlib widget
# 使用外部控件实现动画和交互
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import itertools


fig,ax = plt.subplots()
line, = ax.plot([], [], lw=2)
ax.grid()
xdata, ydata = [], []


def init():# data_gen
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    del xdata[:]#!保证重新开始的时候不会和上一次的线条冲突,不能修改引用
    del ydata[:]
    line.set_data(xdata, ydata)
    ax.figure.canvas.draw()

    return line,


def data_gen():
    for x in np.arange(0, 10*np.pi, 0.1):
        yield x, np.sin(x)


def run(data):
    x, y = data
    xdata.append(x)
    ydata.append(y)
    xmin, xmax = ax.get_xlim()
    if x >= xmax:
        ax.set_xlim(xmin, 2*xmax)
        ax.figure.canvas.draw()
    line.set_data(xdata, ydata)
    return line,


anim = animation.FuncAnimation(fig, run, data_gen, interval=10, init_func=init,blit=True,repeat=False)
plt.show()
anim.save('animation.gif')
plt.close()
```
